# Cleaning GroupStage Tables

1. Using for loop and range to get all the tables and clean the data i.e popping 'Qualification' Column
2. Need to change the matches pld W,D,L etc coloumns values to zero except for Pos and Team Values so for that Have to make a seperate list and then use df.loc[] and columns.isin() method to make the coloumns values 0 except for Team and Pos columns so for that we need to use Bitwise Operator "NOT(~)"
df.loc[:, ~df.columns.isin(columns_to_keep)] -  selects all rows (:) and the columns where the boolean array is True (pld,W,D,L etc..).
= 0 assigns the value 0 to all elements in the selected columns. 


```python
# A -> F 
# 19 - 7*6 + 19 = 61
all_tables =  pd.read_html("https://en.wikipedia.org/wiki/UEFA_Euro_2024#Group_stage")
# Need to change the matches pld W,D,L etc coloumns values to zero except for Pos and Team Values so for that Have to make a seperate list
columns_to_keep = ['Pos', 'Teamvte']
for i in range(19,55,7): 
    print(i)
    df = all_tables[i]
    df.loc[:, ~df.columns.isin(columns_to_keep)] = 0
    df.rename(columns={df.columns[1]:'Team'}, inplace=True)
    df.pop('Qualification')
    print(df)
```

1. Create an Empty dictionary 
2. Use ascii_uppercase as alphabet to name the Groups
3. Use zip() to merge both the data (Alphabets as keys and tables as values)
4. Use pickle (use for serialization and deserialization) to write the file in binary state to dict_table(new file)


```python
all_tables =  pd.read_html("https://en.wikipedia.org/wiki/UEFA_Euro_2024#Group_stage")

dict_table = {} 
columns_to_keep = ['Pos', 'Teamvte']
for letter,i in zip(alphabet, range(19,55,7)): # A=19, B=26 etc...
    df = all_tables[i]
    df.loc[:, ~df.columns.isin(columns_to_keep)] = 0 
    df.rename(columns={df.columns[1]: 'Team'}, inplace=True)
    df.pop('Qualification')
    dict_table[f'Group {letter}'] = df 
    
dict_table.keys()
```

```python
with open('dict_table', 'wb') as output:
    pickle.dump(dict_table, output)
```
